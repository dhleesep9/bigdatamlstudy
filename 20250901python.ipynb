```python
#함수의 효력범위 (local vs global)
a=1
print(a)

def ad_vartest(a):
  a = a + 1

ad_vartest(a)
print(a)
```

```python
#global 글로벌 변수
a=1
print(a)

def ad_vartest():
  global a
  a = a + 1

ad_vartest()
print(a)
```

```python
#lambda 함수: 굳이 권고하지않음 한줄로 쓰기위한 함수 줄바꾸기 귀찮아 별의미없다. 크게 의미두지말자
#일반적 선언형태의 함수
def add1(a,b,c):
  return a+b
#람다 함수
add2 = lambda a,b,c: a+b
add1(1,2,3)
add2(1,2,3)
```

```python
#문자열 띄어쓰기는 쉼표로 쓴다. sep 은 구분자 default 가 띄어쓰기
print("life", "is", "too", "short")
print("life", "is", "too", "short", sep=" ")
print("life", "is", "too", "short", sep="/")
# print(), end 옵션 : end에는 줄바꿈이 숨겨져있다.
print("life", "is", "too", "short", end="\n") #"\n"은 줄바꿈
print("life", "is", "too", "short", end=" ")
print("life", "is", "too", "short")
```

```python
#파일 읽고 쓰기 (파일 입/출력): r w a// w는 무조건 새로만들고 a는 뒤에 추가를한다
f=open("새파일.txt", "w")#a 옵션은 append 한번해보면안다.
f.write("오늘 날씨가 비가 오다가 안오다가...\n")
f.write("라인추가")#생으로는 줄이바뀌지않는다. 줄바꾸고 싶으면 \n 을 써야한다.
f.close()
```

```python
import sys
import random
num_count =int(sys.argv[1])
for i in range(num_count):
  balls=[]
  while len(balls) != 6:
    a=random.randint(1,46)
    if a not in balls:
      balls.append(a)
  print("로또 번호는 {0} 입니다.".format(",".join(map(str, balls))))
#이런식으로 cmd 외부에서 실행가능하다.
```

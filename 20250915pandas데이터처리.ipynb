```python
data = pd.read_csv("/content/weight_log.csv")

print(data.shape)          # 몇개 행row, 몇개 컬럼column
print(data.columns)
print(data.index)
print(data.info()) #자료형 column 같이 많은 정보를 보여준다.
print(data.describe()) #기술 통계 정보를 보여준다.
print(data.describe(percentiles=[0.1,0.9]))
print(data.head(3)) #위에서 부터 몇개만 끊어서 보는식
print(data.tail(3))
```

```python
#조회 방법 (1/2)
# columns 기준 조회 :
data["담당"]
data[["몸무게", "지점"]]#data 옆에는 자료형이 하나밖에못온다 리스트로 강제로 만들어서 넣어주면된다.
col= ["몸무게","지점"]
data[col].tail(3)
```

```python
# index 기준 조회 :
data.loc[3]
row=[3,6,7]
data.loc[row]     #여러개 index 조회
data.loc[3:5]      #index 슬라이싱
```

```python
# index & columns
data.loc[[3,6,7], ["몸무게", "단위"]]
```

```python
#조회방법 2/2
# 조건에 의한 데이터 조회 <- boolean index 방식
data["지점"] == "서초구"
data[data["지점"]=="서초구"] #둘이 똑같은
data.loc[data["지점"]=="서초구", ["담당","지점"]]
#강남구 서초구 사는 사람중에 75 보다 큰사람
#python 에서 and, or -> pandas 에서는 (and ->) &, (or ->) |
cond=((data["지점"]=="강남구") | (data["지점"]=="서초구")) & (data["몸무게"]>75.0)
data[cond]
data.loc[cond,["담당","지점"]]# 특정 컬럼만보고싶을떄 loc를 통해 할수있다.
```

```python
#참고 : by label 방식
data.loc[4:5,["담당", "지점"]]
```

```python
#참고: by position 방식
data.iloc[[4,5],[5,6]]
data.loc[4:5].iloc[0,[5,6]]
```

```python
#missing values 결측치 확인 , 처리 방법 결측치 처리는 의사결정 사항이다.
data.info()
data.isna().sum()
data.isna().sum(axis=1)

# 참고: boolean index 방식
data[data["몸무게"].isna()]
```

```python
# missing values 결측치 처리방법
data["몸무게"].mean() #quantile(), min(), max(), median() 결측치를 메꿀값을 정했다.
data["몸무게"].fillna(75.83)
```

```python
# 컬럼 추가(신규)
data["몸무게(결측처리)"]=data["몸무게"].fillna(75.83)
data
```

```python
#컬럼columns 삭제
data.drop(columns=["몸무게(결측처리)","단위"])# 진짜 지우지는 않고 만약 진짜로 지우고싶다면 data= data.drop(columns=["몸무게(결측처리)","단위"])
data
```
